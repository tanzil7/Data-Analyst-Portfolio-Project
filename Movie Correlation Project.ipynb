```python
import pandas as pd 
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
```

```python
#read the data 
df= pd.read_csv('movies.csv')
df.head()
```

```python
df.tail()
```

```python
 #check if any data is missing in the columns

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} --> {}%'.format(col, pct_missing))
```

```python
for col in df.columns:
    total_missing = df.isnull().sum()
    print('{} --> {}'.format(col, total_missing))
```

```python
df.dtypes
```

```python
df = df.dropna(how='any')
df
```

```python
df['budget'] = df['budget'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df
```

```python
df.sort_values(by=['gross'], ascending=False)
```

```python
#drop any duplicates
df = df.drop_duplicates()
df
```

### I predict that if budget is higher for a movie then the gross is high as well 
### Budget, Companies will have high correlation 


```python
# Scatter plot for Budget vs Gross

plt.title("Budget vs Gross")
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.scatter(x= df['budget'] , y=df['gross'])
plt.show()
```

```python
#Plotting budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":"black"}, line_kws={"color":"red"})
```

```python
#We will use the Pearson 

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation")
plt.xlabel("Movie Feature")
plt.ylabel("Movie Feature")

plt.show()
```

```python
#here going to change change string such as company name, director name etc to numeric value 

df_number = df

for col_name in df_number:
    if (df_number[col_name].dtype == 'object'):
        df_number[col_name]= df_number[col_name].astype('category')
        df_number[col_name] = df_number[col_name].cat.codes
df_number
```

```python
correlation_matrix = df_number.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
sns.set(rc={'figure.figsize': (12, 12)})


plt.title("Correlation")
plt.xlabel("Movie Feature")
plt.ylabel("Movie Feature")

plt.show()
```

```python
correlation_mat=df_number.corr()
corr_pair=correlation_mat.unstack()

corr_pair
```

```python
sort_pairs= corr_pair.sort_values()

high_corr = sort_pairs[(sort_pairs)>0.5 ]
high_corr
```
